```python
# Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt

# Path to the CSV file
file_path = "/content/2022-12-09-enriched.csv"  # Update this with your file's path

# Load the CSV data into a Pandas DataFrame
try:
    cve_data = pd.read_csv(file_path)
    print("Data loaded successfully!")
except FileNotFoundError:
    print("Error: File not found. Please check the file path.")

# Display the first few rows of the dataset to understand its structure
print("Sample data:")
print(cve_data.head())

# Inspect column names to identify the severity field
print("Available columns in the dataset:")
print(cve_data.columns)

# Count vulnerabilities by severity
severity_column = "severity"  # Replace with the correct column name for severity
if severity_column in cve_data.columns:
    severity_counts = cve_data[severity_column].value_counts()
    print("\nNumber of vulnerabilities by severity:")
    print(severity_counts)

    # Plot vulnerabilities by severity
    plt.figure(figsize=(10, 6))
    severity_counts.plot(kind='bar', color='pink', edgecolor='salmon')
    plt.title("Number of Security Vulnerabilities by Severity", fontsize=14)
    plt.xlabel("Severity Level", fontsize=12)
    plt.ylabel("Number of Vulnerabilities", fontsize=12)
    plt.xticks(rotation=45, fontsize=10)
    plt.tight_layout()
    plt.show()
else:
    print(f"Error: '{severity_column}' column not found in the dataset.")
```
